```python
import numpy as np
import matplotlib.pyplot as plt

# --- Simulación con 1,000 dardos ---
print("--- Simulación con 1,000 Dardos ---")
num_dardos_1k = 1000

# 2. Generar coordenadas aleatorias entre 0 y 1.
x_1k = np.random.rand(num_dardos_1k)
y_1k = np.random.rand(num_dardos_1k)

# 3. Determinar si el dardo cayó dentro del círculo.
# Círculo centrado en (0.5, 0.5) con radio 0.5
aciertos_1k = (x_1k - 0.5)**2 + (y_1k - 0.5)**2 < 0.5**2
num_aciertos_1k = np.sum(aciertos_1k)

# 4. Calcular el área aproximada y estimar Pi.
area_estimada_1k = num_aciertos_1k / num_dardos_1k
pi_estimado_1k = area_estimada_1k * 4

# Comparar con los valores exactos.
area_exacta = np.pi / 4
error_1k = abs((pi_estimado_1k - np.pi) / np.pi) * 100

print(f"Fracción de aciertos (Área estimada): {area_estimada_1k:.6f}")
print(f"Área exacta del círculo (π/4): {area_exacta:.6f}")
print(f"Valor estimado de π: {pi_estimado_1k:.6f}")
print(f"Valor real de π: {np.pi:.6f}")
print(f"La aproximación tiene un error de: {error_1k:.4f}%\n")


# --- 5. Repetir con 10,000 dardos ---
print("--- Simulación con 10,000 Dardos ---")
num_dardos_10k = 10000
x_10k = np.random.rand(num_dardos_10k)
y_10k = np.random.rand(num_dardos_10k)
aciertos_10k = (x_10k - 0.5)**2 + (y_10k - 0.5)**2 < 0.5**2
area_estimada_10k = np.sum(aciertos_10k) / num_dardos_10k
pi_estimado_10k = area_estimada_10k * 4
error_10k = abs((pi_estimado_10k - np.pi) / np.pi) * 100

print(f"Fracción de aciertos (Área estimada): {area_estimada_10k:.6f}")
print(f"Valor estimado de π: {pi_estimado_10k:.6f}")
print(f"La aproximación tiene un error de: {error_10k:.4f}%\n")


# --- 5. Repetir con 100,000 dardos ---
print("--- Simulación con 100,000 Dardos ---")
num_dardos_100k = 100000
x_100k = np.random.rand(num_dardos_100k)
y_100k = np.random.rand(num_dardos_100k)
aciertos_100k = (x_100k - 0.5)**2 + (y_100k - 0.5)**2 < 0.5**2
area_estimada_100k = np.sum(aciertos_100k) / num_dardos_100k
pi_estimado_100k = area_estimada_100k * 4
error_100k = abs((pi_estimado_100k - np.pi) / np.pi) * 100

print(f"Fracción de aciertos (Área estimada): {area_estimada_100k:.6f}")
print(f"Valor estimado de π: {pi_estimado_100k:.6f}")
print(f"La aproximación tiene un error de: {error_100k:.4f}%\n")


# --- Visualización de la simulación con 1,000 dardos ---
print("--- Visualizando la simulación de 1,000 Dardos ---")
fig, ax = plt.subplots(figsize=(8, 8))

# Dibujar los dardos que cayeron fuera (fallos)
ax.scatter(x_1k[~aciertos_1k], y_1k[~aciertos_1k], color='blue', label='Fallo', s=5)
# Dibujar los dardos que cayeron dentro (aciertos)
ax.scatter(x_1k[aciertos_1k], y_1k[aciertos_1k], color='red', label='Acierto', s=5)

# Dibujar el borde del círculo
circulo = plt.Circle((0.5, 0.5), 0.5, color='black', fill=False, linewidth=2, label='Borde del Círculo')
ax.add_artist(circulo)

ax.set_title('Simulación de 1,000 Dardos')
ax.set_xlabel('Coordenada X')
ax.set_ylabel('Coordenada Y')
ax.set_aspect('equal', adjustable='box')
ax.legend()
plt.grid(True)
plt.show()
```
