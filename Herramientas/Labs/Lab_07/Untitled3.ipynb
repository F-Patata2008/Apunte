### Desarrollo

Se realizan 3 simulaciones con **1.000, 10.000 y 100.000** dardos. Para cada una:
1.  Se generan dos vectores `x` e `y` con coordenadas aleatorias entre 0 y 1.
2.  Se determina si cada punto (dardo) `(x, y)` cae dentro del círculo con la condición `(x - 0.5)² + (y - 0.5)² < 0.5²`.
3.  Se calcula el área estimada (proporción de aciertos) y se usa para aproximar **π**.
4.  Se calcula el error de la aproximación comparado con el valor real de **π**.

### Conclusiones

Al aumentar el número de dardos (`n`), la aproximación de **π** se vuelve más precisa y el error porcentual disminuye. Con 100,000 dardos, la estimación es bastante cercana al valor real. Aunque para resolver problemas, sigo prefiriendo C++ con Neovim.

---
_A continuación, se presenta el código y los resultados de las simulaciones._

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Simulación con n = 1,000 ---
print("--- Simulación con n = 1,000 ---")
n = 1000
x = np.random.rand(n)
y = np.random.rand(n)

# Comprobar si están dentro del círculo
si = (x - 0.5)**2 + (y - 0.5)**2 < 0.5**2
nsi = np.sum(si)

# Calcular área y pi
aprox_area = nsi / n
aprox_pi = aprox_area * 4
error = abs((aprox_pi - np.pi) / np.pi) * 100

print(f"Área estimada: {aprox_area:.6f}")
print(f"Pi estimado: {aprox_pi:.6f}")
print(f"Error porcentual: {error:.4f}%\n")


# --- Simulación con n = 10,000 ---
print("--- Simulación con n = 10,000 ---")
n = 10000
x = np.random.rand(n)
y = np.random.rand(n)

si = (x - 0.5)**2 + (y - 0.5)**2 < 0.5**2
nsi = np.sum(si)

aprox_area = nsi / n
aprox_pi = aprox_area * 4
error = abs((aprox_pi - np.pi) / np.pi) * 100

print(f"Área estimada: {aprox_area:.6f}")
print(f"Pi estimado: {aprox_pi:.6f}")
print(f"Error porcentual: {error:.4f}%\n")


# --- Simulación con n = 100,000 ---
print("--- Simulación con n = 100,000 ---")
n = 100000
x = np.random.rand(n)
y = np.random.rand(n)

si = (x - 0.5)**2 + (y - 0.5)**2 < 0.5**2
nsi = np.sum(si)

aprox_area = nsi / n
aprox_pi = aprox_area * 4
error = abs((aprox_pi - np.pi) / np.pi) * 100

print(f"Área estimada: {aprox_area:.6f}")
print(f"Pi estimado: {aprox_pi:.6f}")
print(f"Error porcentual: {error:.4f}%\n")
```
