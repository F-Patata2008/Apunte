
# Introducción a operaciones básicas en Python usando Google Colab

En este notebook exploraremos operaciones fundamentales en Python: operaciones entre vectores y escalares, operaciones entre vectores, y la creación de gráficos simples usando Google Colab.


## Creación de vectores

Primero crearemos dos vectores de números reales aleatorios.


```python
import numpy as np

# Crear vectores aleatorios
x = np.random.rand(10)
y = np.random.rand(10)

print("Vector x:", x)
print("Vector y:", y)
```


## Operaciones entre vectores y escalares

Multiplicaremos el vector por un escalar, y también sumaremos y restaremos un escalar a cada elemento del vector.


```python
# Multiplicación por escalar
x2 = 2 * x
print("x multiplicado por 2:", x2)

# Suma de un escalar
x_mas_1 = x + 1
print("x más 1:", x_mas_1)

# Resta de un escalar
x_menos_1 = x - 1
print("x menos 1:", x_menos_1)
```


## Operaciones entre vectores

Veamos qué sucede si sumamos, restamos o multiplicamos dos vectores del mismo tamaño.


```python
# Suma de vectores
suma = x + y
print("Suma x + y:", suma)

# Resta de vectores
resta = x - y
print("Resta x - y:", resta)

# Producto elemento a elemento
producto = x * y
print("Producto elemento a elemento:", producto)
```


## Otros tipos de operaciones con vectores

También se pueden realizar operaciones como divisiones elemento a elemento, potencias, y más.


```python
# División elemento a elemento
division = x / y
print("División x / y:", division)

# Elevar elementos al cuadrado
cuadrado = x**2
print("x al cuadrado:", cuadrado)
```


## Gráficos básicos en Python

Finalmente, graficaremos los vectores para visualizar los datos.


```python
import matplotlib.pyplot as plt
import math

# Gráfico de dispersión de x vs y
plt.figure(figsize=(6,6))
plt.scatter(x, y, color='blue', label='Puntos (x,y)')
plt.title('Gráfico de dispersión entre x e y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Gráfico de lineas de a vs sin(a)
a = np.linspace(0, 2* math.pi, 100)
sin = np.vectorize(math.sin)
b = sin(a)
plt.plot(a,b)
```


## Resumen

- Operar vectores con escalares modifica cada elemento individualmente.
- Operar vectores entre sí aplica la operación elemento a elemento (suma, resta, multiplicación, división).
- Es posible graficar relaciones entre vectores de manera sencilla usando 	matplotlib.
